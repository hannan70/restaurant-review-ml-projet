```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=['the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
# define the vocabulary size
voc_size=10000
```

```python
one_hot_repre = [one_hot(word, voc_size) for word in sent]
one_hot_repre
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
max_length = 10
embedding_docs = pad_sequences(one_hot_repre, padding="pre", maxlen=max_length)
embedding_docs
```

```python
dim=10

model = Sequential()
model.add(Embedding(input_dim=voc_size, output_dim=dim, input_length=max_length))
model.compile("adam", "mse")
```

```python
model.summary()
```

```python
model.predict(embedding_docs)
```
